```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
iris=pd.read_csv('iris.csv')
```

```python
x1=iris.loc[iris['class']=='Iris-setosa']
```

```python
x2=iris.loc[iris['class']=='Iris-virginica']
```

```python
x3=iris.loc[iris['class']=='Iris-versicolor']
```

```python
plt.figure()
plt.scatter(x1.iloc[:,2],x1.iloc[:,3],marker='o')
plt.scatter(x2.iloc[:,2],x2.iloc[:,3],marker='v')
plt.scatter(x3.iloc[:,2],x3.iloc[:,3],marker='^')
```

```python
sns.barplot(x='class',y='pl',data=iris)
```

```python
sns.lineplot(x='class',y='pl',data=iris)
```

```python
iris['class'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
plt.hist(iris['sl'])
plt.xlabel("sepal length")
plt.ylabel("count")
```
